### 7.2 Analyze the monthly sum of confirmed cases, deaths and recoveries for countries to understand the progression of the pandemic from the merged dataset. 

```python
merge_df.head(2)
```

```python
merge_df['Date'].dtypes
```

```python
merge_df = merge_df.sort_values(['Country/Region','Date'])

merge_df['daily_confirmed'] = merge_df.groupby('Country/Region')['confirmed'].diff()
merge_df['daily_deaths'] = merge_df.groupby('Country/Region')['deaths'].diff()
merge_df['daily_recovered'] = merge_df.groupby('Country/Region')['cases'].diff()
```

```python
merge_df[['daily_confirmed','daily_deaths','daily_recovered']] = \
merge_df[['daily_confirmed','daily_deaths','daily_recovered']].fillna(0)

merge_df = merge_df[
    (merge_df['daily_confirmed'] >= 0) &
    (merge_df['daily_deaths'] >= 0) &
    (merge_df['daily_recovered'] >= 0)
]
```

```python
merge_df['Year_Month'] = merge_df['Date'].dt.to_period('M')
```

```python
monthly_data = merge_df.groupby(
    ['Country/Region','Year_Month']
)[['daily_confirmed','daily_deaths','daily_recovered']].sum().reset_index()
```

```python
monthly_data.sort_values('daily_confirmed', ascending=False).head(10)
```
